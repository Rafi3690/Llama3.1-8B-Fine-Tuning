```python

import pandas as pd
from sklearn.model_selection import train_test_split
from datasets import Dataset

class DatasetProcessor:
    def __init__(self, csv_path):
        self.df = pd.read_csv(csv_path)

    def preprocess(self):
        self.df = self.df.dropna(
            subset=["Topics", "Questions", "Answers"]
        )

        # Remove rare topics
        topic_counts = self.df["Topics"].value_counts()
        valid_topics = topic_counts[topic_counts >= 2].index
        self.df = self.df[self.df["Topics"].isin(valid_topics)]

        self.df["input_text"] = (
            "প্রসঙ্গ: " + self.df["Topics"].astype(str) + "\n"
            "প্রশ্ন: " + self.df["Questions"].astype(str).str.strip()
        )

        self.df["target_text"] = self.df["Answers"].astype(str).str.strip()

        train_df, val_df = train_test_split(
            self.df,
            test_size=0.01,
            random_state=42,
            stratify=self.df["Topics"]
        )

        train_dataset = Dataset.from_pandas(
            train_df[["input_text", "target_text"]].reset_index(drop=True)
        )
        val_dataset = Dataset.from_pandas(
            val_df[["input_text", "target_text"]].reset_index(drop=True)
        )

        return train_dataset, val_dataset
```

```python
processor = DatasetProcessor(
    "/kaggle/input/bangla-emotion/BengaliEmpatheticConversationsCorpus .csv"
)

train_dataset, val_dataset = processor.preprocess()
```

```python
PROMPT_TEMPLATE = """### System:
আপনি একজন অভিজ্ঞ, সহানুভূতিশীল এবং দায়িত্বশীল মানসিক স্বাস্থ্য সহায়ক।
আপনি বিচার করবেন না, ভয় দেখাবেন না, এবং ক্ষতিকর বা বিপজ্জনক পরামর্শ দেবেন না।
আপনি সংক্ষিপ্ত, স্পষ্ট এবং সহমর্মিতাপূর্ণ উত্তর দেবেন।

### User:
{input}

### Assistant:
"""
```

```python
def format_prompts(examples):
    texts = []
    for inp, out in zip(examples["input_text"], examples["target_text"]):
        texts.append(
            PROMPT_TEMPLATE.format(
                input=inp,
                output=out
            )
        )
    return {"text": texts}
```

```python
train_dataset = train_dataset.map(
    format_prompts,
    batched=True,
    remove_columns=train_dataset.column_names
)

val_dataset = val_dataset.map(
    format_prompts,
    batched=True,
    remove_columns=val_dataset.column_names
)
```

```python
from unsloth import FastLanguageModel
import torch

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name="unsloth/Meta-Llama-3.1-8B-Instruct",
    dtype=None,
    load_in_4bit=True,
)
```

```python
model = FastLanguageModel.get_peft_model(
    model,
    r=8,
    target_modules=[
        "q_proj", "k_proj", "v_proj",
        "o_proj", "gate_proj",
        "up_proj", "down_proj"
    ],
    lora_alpha=16,
    lora_dropout=0.05,
    bias="none",
    use_gradient_checkpointing=True,
)
```

```python
from trl import SFTTrainer
from transformers import TrainingArguments

trainer = SFTTrainer(
    model=model,
    tokenizer=tokenizer,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    dataset_text_field="text",
    packing=False,
    args=TrainingArguments(
        per_device_train_batch_size=2,
        gradient_accumulation_steps=4,
        warmup_steps=40,
        max_steps=10,
        learning_rate=2e-4,
        fp16=not torch.cuda.is_bf16_supported(),
        bf16=torch.cuda.is_bf16_supported(),
        logging_steps=20,
        eval_strategy="steps",
        eval_steps=1,
        save_steps=10,
        output_dir="./unsloth-bangla-llama",
        optim="adamw_8bit",
        report_to="none",
        # logging_strategy="steps",

    ),
)
```

```python
trainer.train()
```

```python
eval_results = trainer.evaluate()
print(eval_results)
```

```python
import torch
from nltk.translate.bleu_score import corpus_bleu, SmoothingFunction
from tqdm import tqdm
import nltk
nltk.download("punkt")
```

```python
def extract_assistant_response(text):
    if "### Assistant:" in text:
        return text.split("### Assistant:")[-1].strip()
    return text.strip()
```

```python
def generate_response(model, tokenizer, prompt, max_new_tokens=256):
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=max_new_tokens,
            do_sample=True,
            temperature=0.7,
            top_p=0.9,
            repetition_penalty=1.1,
        )

    decoded = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return extract_assistant_response(decoded)
```

```python
def compute_bleu(model, tokenizer, eval_dataset, num_samples=50):
    predictions = []
    references = []

    smooth = SmoothingFunction().method4

    for i in tqdm(range(min(num_samples, len(eval_dataset)))):
        example = eval_dataset[i]

        prompt = example["text"]              
        reference = example["target_text"]   

        prediction = generate_response(model, tokenizer, prompt)

        if len(prediction.strip()) == 0:
            continue

        predictions.append(prediction.split())
        references.append([reference.split()])

    bleu = corpus_bleu(
        references,
        predictions,
        smoothing_function=smooth
    )

    return bleu
```
